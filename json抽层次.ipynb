```python
import io
import pandas as pd
import re,json
with open('LogExport_11_2017-06-13_2017-06-17.csv',encoding = 'gb18030',) as file:
    data = file.read()
pd.set_option('max_colwidth',800)
df = pd.read_csv(io.StringIO(data),sep = '\t')
df.head()
```

```python
df.columns = list(map(str.strip,df.columns.values))
df.columns
```

```python
def get_output(arow):
    df = pd.DataFrame(arow['Output'])
    df['companyId'] = arow['Input']['companyId']
    df['userId'] = arow['Input']['userId']
    df['TimeStamp'] = arow['TimeStamp']
    return df
```

```python
dfn = df[['TimeStamp']].copy()
dfn['Input'] = df.Input.map(json.loads)
dfn['Output'] = df.Output.map(json.loads)
frames = []
for i in range(len(df)):
    frames.append(get_output(dfn.iloc[i,:]))
result = pd.concat(frames,axis = 0,ignore_index = True)
result
```
